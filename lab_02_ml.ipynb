```python
def get_av_grade_by_sex(sex, subject):
    shit = df.loc[df.SEXTYPENAME == sex]
    return shit[subject].mean()


def compare_results_by_sex(subjects, labels):
    fig = plt.figure(figsize=(25,12))
    ax = fig.add_subplot(111)
    n = len(subjects)
    ind = np.arange(n) *2
    width = 0.3
    man = 'чоловіча'
    woman = 'жіноча'

    man_values = [get_av_grade_by_sex(man, sbj) for sbj in subjects]
    woman_values = [get_av_grade_by_sex(woman, sbj) for sbj in subjects]

    man_rects = ax.bar(ind, man_values, width, color='b')
    woman_rects = ax.bar(ind+width, woman_values, width, color='r')

    ax.set_ylabel('Бали')
    ax.set_xticks(ind+width)
    ax.set_xticklabels(labels)
    ax.legend( (man_rects[0], woman_rects[0]), (man, woman) )

    def autolabel(rects):
        for rect in rects:
            h = rect.get_height()
            ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h), ha='center', va='bottom')

    autolabel(man_rects)
    autolabel(woman_rects)

    plt.show()

subjects = [col for col in df.columns if col.endswith('Ball100')]
labels = [df[col].mode()[0] for col in df.columns if col.endswith('Test')]
compare_results_by_sex(subjects, labels)
```

Можна зробити припущення, що ті, хто здавали математику, також гарно здавали фізику чи хімію

```python
def lin(sub1, sub2):
    ss = df[[sub1, sub2]].replace({0:np.nan}).dropna()
    b1, b2 = [], []
    for ball in sorted(set(ss[sub1])):
        people = ss[ss[sub1] == ball]
        mean = people[sub2].mean()
        b1.append(ball)
        b2.append(mean)
    return pandas.DataFrame( {sub1: b1, sub2: b2})

s1, s2 = "physBall100", "mathBall100"
data = lin(s1, s2)
sns.lmplot(s1, s2, data)
plt.show()
```

## PLA


```python
kek = df[['mathBall100', 'UkrBall100', 'histBall100', 'ClassProfileNAME']]
kek = kek.dropna()
kek = kek.loc[kek['ClassProfileNAME'].isin(['Історичний', 'Математичний'])]
print(len(kek))
#print(len(kek.loc[kek['ClassProfileNAME'].isin(['Фізико-математичний'])]))

only_hist = kek.loc[kek['ClassProfileNAME'].isin(['Історичний'])]
only_math = kek.loc[kek['ClassProfileNAME'].isin(['Математичний'])]
print(len(only_hist))
print(len(only_math))

kek.head(100)
```

```python
kek['ClassProfileNAME'] = np.where(kek.iloc[:, -1]=='Історичний', 0, 1)
data = np.asarray(kek, dtype = 'float64')
x_data = data[:,:3]
y_data = data[:,-1]
print('x_data shape', x_data.shape)
print('y_data shape', y_data.shape)
```

```python
sns.set(style = "darkgrid")

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection = '3d')


ax.set_xlabel("mathBall100")
ax.set_ylabel("UkrBall100")
ax.set_zlabel("histBall100")

ax.scatter(only_hist['mathBall100'], only_hist['UkrBall100'], only_hist['histBall100'], color='r')
ax.scatter(only_math['mathBall100'], only_math['UkrBall100'], only_math['histBall100'], color='b')

plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data,y_data, test_size = 0.1, stratify = y_data, random_state = 1)
print('shapes: ')
print('x_train shape: ', x_train.shape)
print('x_test shape: ', x_test.shape)
print('y_train shape: ', y_train.shape)
print('y_test shape: ', y_test.shape)
```

```python
from sklearn.metrics import accuracy_score


class PLA:
    def __init__ (self):
        self.weights = None
        self.b = None

    def model(self, x):
        return 1 if (np.dot(self.weights, x) >= self.b) else 0

    def predict(self, X):
        Y = []
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.array(Y)

    def fit(self, X, Y, epochs = 1, lr = 1):
        self.weights = np.ones(X.shape[1])
        self.b = 0
        accuracy = {}
        max_accuracy = 0
        weights_matrix = []
        for i in range(epochs):
            for x, y in zip(X, Y):
                y_pred = self.model(x)
                if y == 1 and y_pred == 0:
                    self.weights = self.weights + lr * x
                    self.b = self.b - lr * 1
                elif y == 0 and y_pred == 1:
                    self.weights = self.weights - lr * x
                    self.b = self.b + lr * 1

            weights_matrix.append(self.weights)
            accuracy[i] = accuracy_score(self.predict(X), Y)
            if (accuracy[i] > max_accuracy):
                max_accuracy = accuracy[i]
                chkptw = self.weights
                chkptb = self.b
        self.weights = chkptw
        self.b = chkptb
        print(max_accuracy)
        fig2 = plt.figure(figsize=(10,10))
        ax2 = fig2.gca()
        ax2.plot(list(accuracy.keys()), list(accuracy.values()))
        ax2.set_xlabel("Epoch #")
        ax2.set_ylabel("Accuracy")
        plt.show()
        return np.array(weights_matrix)
```

```python
perceptron = PLA()
wt_matrix = perceptron.fit(x_train, y_train, 1000, 0.1)
print(len(wt_matrix))
#making predictions on test data
Y_pred_test = perceptron.predict(x_test)

#checking the accuracy of the model
print(accuracy_score(Y_pred_test, y_test))
```
